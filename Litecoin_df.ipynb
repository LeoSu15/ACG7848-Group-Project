```python
L_age= pd.read_csv("https://raw.githubusercontent.com/LeoSu15/ACG7848-Group-Project/main/Litecoin_age.csv")
L_dev=pd.read_csv("https://raw.githubusercontent.com/LeoSu15/ACG7848-Group-Project/main/Litecoin_dve.csv")
L_mar=pd.read_csv("https://raw.githubusercontent.com/LeoSu15/ACG7848-Group-Project/main/Litecoin_MarketCap.csv")
L_social=pd.read_csv("https://raw.githubusercontent.com/LeoSu15/ACG7848-Group-Project/main/Litecoin_social.csv")
L_volume=pd.read_csv("https://raw.githubusercontent.com/LeoSu15/ACG7848-Group-Project/main/Litecoin_volumes.csv")
L_return=pd.read_csv("https://raw.githubusercontent.com/LeoSu15/ACG7848-Group-Project/main/Litecoin_daily_return.csv")
df=L_return
```

```python
L_social.dropna(inplace=True)
L_dev.dropna(inplace=True)
L_mar.dropna(inplace=True)
L_age.dropna(inplace=True)
L_return.dropna(inplace=True)
#b_volume.drop(['Social Volume'], axis='columns', inplace=True)
L_volume.dropna(inplace=True)
```

```python
finaldf = pd.merge(L_age,L_dev, how = "left", on = ['Date'])
finaldf = pd.merge(finaldf,L_mar, how = "left", on = ['Date'])
finaldf = pd.merge(finaldf,L_social, how = "left", on = ['Date'])
finaldf = pd.merge(finaldf,L_volume, how = "left", on = ['Date'])

```

```python
finaldf = pd.merge(L_return,finaldf, how = "left", on = ['Date'])
```
