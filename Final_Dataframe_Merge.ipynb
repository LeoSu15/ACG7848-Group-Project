```python
a = pd.read_pickle(path + "dailyreturns_df.pkl")
b = pd.read_pickle(path + "marketcap_usd_daily.pkl")
c = pd.read_pickle(path + "daily_trading_volume_usd_daily.pkl")
d = pd.read_pickle(path + "dev_activity_daily.pkl")
finaldf = pd.merge(a,b, how = "left", on = ['Coin', 'datetime'])
finaldf = pd.merge(finaldf, c, how = "left", on = ['Coin', 'datetime'])
```

```python
namelist = os.listdir(path)
for i in namelist:
    if i != "dailyreturns_df.pkl":
        try:
            temp = pd.read_pickle(path+i)
            finaldf = pd.merge(finaldf, temp, how = "left", on = ['Coin','datetime'])
        except:
            print(i)
            pass
```
